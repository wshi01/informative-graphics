Now for (2) finding the number of countries that use each vaccine. 

```python
df_2 = pd.read_excel('D:/CCNY/Spring 2023/CSc 47400/informative-graphics/COVID-19 Data Analysis/1/Data/vaccination-data-WHO-12-28-2022-NO-DETAILS for-vaciinetype-date.xlsx')
df_2['VACCINES_USED'] = df_2['VACCINES_USED'].apply(lambda x: x.split(','))
vaccines_used = df_2['VACCINES_USED']
vaccines_used
vaccine_co = vaccines_dict

def plot_choropleth(name):
    counter = 0 
    lst = []
    for i in vaccines_used:
        if name in i:
            lst.append(counter)
            counter += 1
    df = df_2.iloc[lst]
    vaccine_co[name] = counter
    fig = px.choropleth(df, locations='ISO3', color='ISO3',
                        title='All countries that use ' + name + " with a total of " + str(counter) + " countries",
                        range_color=(1, 1),
                        color_discrete_sequence=['green'])
    fig.update_geos(fitbounds='locations')
    fig.show()

for i in vaccines_dict: 
    plot_choropleth(i)

print(vaccine_co)
```

```python
import pandas as pd
import plotly.graph_objects as go
vaccine_counts = pd.DataFrame({'VACCINE_TYPE': vaccine_co.keys(),
                               'COUNTRY_COUNT': vaccine_co.values()})

fig = go.Figure()
fig.add_trace(go.Scatter(
    y=vaccine_counts['COUNTRY_COUNT'],
    x=vaccine_counts['VACCINE_TYPE'],
    
    mode='markers',
    marker=dict(
        size=vaccine_counts['COUNTRY_COUNT'],
        sizemode='area',
        sizeref=0.1,
        color=vaccine_counts['COUNTRY_COUNT'],
        colorscale='Viridis'
    ),
    text=vaccine_counts['VACCINE_TYPE']
))
fig.update_layout(
    title='Number of countries using each vaccine',
    xaxis_title='Vaccine Type',
    yaxis_title='Number of Countries'
)
fig.show()
```
